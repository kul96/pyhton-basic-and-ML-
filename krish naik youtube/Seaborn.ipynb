#### Correlation with Heatmap
* correlation value range [-1,1]
* correlation b/w only int,float type not in object type
* correaltion mean = find interrelationship
* A correlation heatmap uses cells, typically in a monochromatic scale, to show a 2D correlation matrix(tabel) between two discrete dimensions or event types. It is very important in Feature Selection

```python
df.corr() # only realation b/w float and int
```

```python
sns.heatmap(df.corr())
```

### JoinPlot
A join plot allows to study the realtionship between 2 numeric varaibles. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df) #bydefault kind = 'scatter'
sns.jointplot(x='tip',y='total_bill',data=df,color='g',kind="hex")
sns.jointplot(x='tip',y='total_bill',data=df,color='r',kind="reg")
sns.jointplot(x='tip',y='total_bill',data=df,color='m',kind="resid")
sns.jointplot(x='tip',y='total_bill',data=df,color='b',kind="kde")

# kind : { "scatter" | "reg" | "resid" | "kde" | "hex" },bydefault = scatter
```

### Pair plot
A 'pair plot' is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaboration on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

### Dist plot
Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip']) 
#bydefault kde='true',kde = karnel density estimation
```

```python
sns.distplot(df['tip'],kde=False)
# x-axis = tip value
# y-axis = show count (if kde present then show density)
```

## Categorical feature plots(object type)
 Seaborn also helps us in doing the analysis on Categorial Data points. In this section we will discuss about 
 1. countplot
 2. bar plot
 3. boxplot
 4. violinplot

#### 1. Count plot
ony one axis value you have to mention in graph

```python
sns.countplot(df['sex'])
# or
sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('time',data=df)
```

```python
sns.countplot(y='day',data=df)
```

#### 2. bar plot
you have to mention x,y both value

```python
sns.barplot(x='tip',y='sex',data=df)
```

```python
sns.barplot(x='total_bill',y='day',data=df)
```

```python
df.head()
```

#### 3. Box plot
A box and whisker plot(sometime called a boxplot)is a graph that presents information from a five-number summary

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
# categorize my data based on some other catergories
sns.boxplot(x='total_bill',y='day',data=df,hue='sex')
```

#### 4. violin plot
violin plot helps us to see both the distribution of data in terms of kernal density estimation(KDE) and the box plot

```python
sns.violinplot(y='day',x='total_bill',data=df)
```
