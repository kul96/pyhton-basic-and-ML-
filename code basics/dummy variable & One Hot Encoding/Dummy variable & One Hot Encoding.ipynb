```python
import pandas as pd
df = pd.read_csv('homeprices.csv')
df
```

```python
## add dummy variable 
## panda give dummy method
dummies = pd.get_dummies(df['town'])
print(dummies)
# df.append(dummies)
```

```python
## concat column
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
## if total 'n' dummy variable created then you have to drop '1' dummy column 
# because it causes problem in machine learnig model
final = merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
## x variable or traning data except 'price'
x = final.drop('price',axis='columns')
x
```

```python
# label of model 'price'
y = final['price']
y
```

```python
model.fit(x,y)
```

```python
## now check how much my model is accurate
print(x)
print(y)
print(x.shape)
print(y.shape)
model.score(x,y) # 0.95 show 95% work correctly
```
