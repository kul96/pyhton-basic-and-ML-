```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['age'],df['bought_insurance'],c='r')

# linear regression 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(df[['age']],df['bought_insurance'])
plt.plot(df[['age']],model.predict(df[['age']]))
## reshape array to 2D
# x = x.reshape(-,1)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['age'],df['bought_insurance'],c='r',marker='+')

## logistic regression
from sklearn.linear_model import LogisticRegression

# model1 = LogisticRegression(C=1.0, solver='lbfgs')
model1 = LogisticRegression()
model1.fit(df[['age']],df['bought_insurance'])
plt.plot(df['age'],model1.predict(df[['age']]))
```

```python
## create test and training data 
from sklearn.model_selection import train_test_split
```

```python
### always about 'x' that is always in 2D 
x_train,x_test,y_train,y_test = train_test_split(df[['age']],df['bought_insurance'],test_size=0.2)
```

```python
model2 = LogisticRegression(solver='lbfgs')
```

```python
model2.fit(x_train,y_train)
```

```python
model2.predict(x_test)
```

```python
model2.score(x_test,y_test)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['age'],df['bought_insurance'],c='r',marker='+')

plt.plot(x_test,model2.predict(x_test))
```
