```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = sns.load_dataset('flights')
a.head()
```

```python
plt.bar(x=a['month'],height=a['passengers'])
```

## why use seaborn if matplotlib already present
 to show more meaningfull and beautiful

```python
sns.relplot(x='passengers',y='month',hue='year',data=a) # hue for 3d represent
```

```python
b = sns.load_dataset('tips')
print(b.head())
sns.relplot(x="time",y="tip",data=b,kind='line')
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',data=b,)
sns.catplot(x='day',y='total_bill',data=b,)
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind='point')
sns.catplot(x='day',y='total_bill',data=b,kind='bar')
sns.catplot(x='day',y='total_bill',data=b,kind='swarm')
sns.catplot(x='day',y='total_bill',data=b,kind='box')
sns.catplot(x='day',y='total_bill',data=b,kind='violin')
sns.catplot(x='day',y='total_bill',data=b,hue='sex',kind='boxen')

# kind : string, optional
#     The kind of plot to draw (corresponds to the name of a categorical
#     plotting function. Options are: "point", "bar", "strip", "swarm",
#     "box", "violin", or "boxen"
```
